```python
import requests
from bs4 import BeautifulSoup
```

```python
recipes_url = "https://www.allrecipes.com/recipes/"
recipes_response = requests.get(recipes_url)
recipes_soul = BeautifulSoup(recipes_response.text, "html.parser")
```

```python
recipe_list = []
for recipe in recipes_soul.find_all("a", {"class": "carouselNav__link recipeCarousel__link"}, href = True):
    recipe_list.append(recipe["href"])
```

```python
with open("recipes.txt", "w") as file:
    for i in range(len(recipe_list)):
        recipe_appetizers_response = requests.get(recipe_list[i])
        recipe_appetizers_soul = BeautifulSoup(recipe_appetizers_response.text, "html.parser")
        recipe_appetizers_list = []
        category = ""
        for recipe in recipe_appetizers_soul.find_all("h1", {"class": "category-page-heading"}):
            if recipe:
                category = recipe.text
        for recipe in recipe_appetizers_soul.find_all("div", {"class": "card__detailsContainer-left"}):
          if recipe.a:
              recipe_appetizers_list.append(recipe.a["href"])
        for k in range(len(recipe_appetizers_list)):
            file.write(recipe_appetizers_list[k] + " | " + category + " | ")
            recipe_strawberry_response = requests.get(recipe_appetizers_list[k])
            recipe_strawberry_soul = BeautifulSoup(recipe_strawberry_response.text, "html.parser")
            for recipe in recipe_strawberry_soul.find_all("h1", {"class": "headline heading-content"}):
                if recipe:
                    file.write(recipe.text + " | ")
            for ingredient in recipe_strawberry_soul.find_all("span", {"class": "ingredients-item-name"}):
                file.write(ingredient.text + " - ")
            file.write(" | ")
            for step in recipe_strawberry_soul.find_all("li", {"class": "subcontainer instructions-section-item"}):
                for direction in step.find_all("div", {"class": "paragraph"}):
                    if direction.p:
                        file.write(direction.p.text + " ")
            file.write("\n")
file.close()
```
